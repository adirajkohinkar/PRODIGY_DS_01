```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np 
import os
```

```python
df=pd.read_csv("ab36b185-bbba-406a-976e-e3e032da6d67_Data.csv")
```

```python
df.head()
```

```python
df_final=df.drop(['Series Code','Country Code','Series Name'] ,axis=1)
```

```python
chosen_country = 'Algeria'
country_data = df_final[df_final['Country Name'] == chosen_country]
```

```python


years = country_data.columns[2:]
population = country_data.iloc[:, 2:].values.flatten()

plt.figure(figsize=(8,8))
plt.bar(years, population, color='skyblue')
plt.xlabel('Year')
plt.ylabel('Population')
plt.title(f'Population of {chosen_country} Over the Years')


plt.show()
     
```

```python
plt.hist(population, bins=10, color='skyblue', edgecolor='black')
plt.show()
```
